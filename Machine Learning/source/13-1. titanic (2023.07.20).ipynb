```python
import warnings
warnings.filterwarnings(action = 'ignore')
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestClassifier
```

```python
gs=pd.read_csv('../data/titanic/gender_submission.csv')
```

```python
test_data = pd.read_csv('../data/titanic/test.csv')
```

```python
test_data
```

```python
train_data = pd.read_csv('../data/titanic/train.csv')
```

```python
tmp = []
for each in train_data['Sex']:
    if each =='female':
        tmp.append(0)
    elif each =='male':
        tmp.append(1)
    else:
        tmp.append(np.nan)
```

```python
train_data['Sex']=tmp
train_data.head()
```

```python
tmp = []
for each in test_data['Sex']:
    if each =='female':
        tmp.append(0)
    elif each =='male':
        tmp.append(1)
    else:
        tmp.append(np.nan)
            
test_data['Sex']=tmp
test_data.head()          
```

```python
train_data = train_data [train_data['Age'].notnull()]
train_data = train_data [train_data['SibSp'].notnull()]
train_data = train_data [train_data['Parch'].notnull()]
train_data = train_data [train_data['Fare'].notnull()]
train_data.info()
```

```python
test_data.info()
```

```python
train_pre= train_data[['Pclass','Sex','SibSp','Parch']]
train_pre.head()
```

```python
test_pre= test_data[['Pclass','Sex','SibSp','Parch']]
test_pre.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test=train_test_split(train_pre,train_data[['Survived']], test_size=0.2,random_state=13)
```

```python
from sklearn.ensemble import RandomForestClassifier
Ran_clf= RandomForestClassifier(n_estimators=13)
Ran_clf.fit(X_train,y_train)
print('Score:{}'.format(Ran_clf.score(X_train,y_train)))
```

```python
submission = pd.DataFrame({'PassengerId':test_data['PassengerId'],'Survived':Ran_clf.predict(test_pre)})
```
