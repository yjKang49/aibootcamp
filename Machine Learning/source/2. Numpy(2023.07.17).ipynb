## concatenate
- Numpy array  

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
np.vstack((a, b))
```

```python
a = np.array([ [1], [2], [3] ])
b = np.array([ [4], [5], [6] ])
np.hstack((a, b))
```

```python
a = np.array([[1, 2, 3]])
b = np.array([[4, 5, 6]])
np.concatenate((a,b), axis = 0)
```

```python
a = np.array([ [1, 2], [3, 4] ])
b = np.array([ [5, 6] ])
np.concatenate((a, b.T), axis = 1)  # a.T a 
# .T   
print(b)
print(b.T)
```
