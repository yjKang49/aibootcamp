### 5. 피마 인디언 당뇨병 예측 실행

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X = df.iloc[:, 0:8]
# 당뇨병 여부 Y
y = df.iloc[:, 8]
```

```python
# 모델을 설정한다. 모델에서 최초의 입력(input_dim)과 최종 출력( 1 )은 고정. 그 외의 hidden layer에서 압*출력값은 개발자 역량
model = Sequential()
model.add(Dense(12, input_dim = 8, activation = 'relu', name = 'Dense_1'))      #입력 8개. 바뀌면x. hidden layer 에 output 12개 전달
model.add(Dense(8, activation = 'relu', name = 'Dense_2'))      # 다시 입력이 12개, 출력이 8개
model.add(Dense(1, activation = 'sigmoid', name = 'Dense_3'))   # 입력이 8개, 출력이 1개. 
model.summary()
```

```python
# 모델 컴파일
# model.compile 부분은 앞서 지정한 모델이 효과적으로 구현될 수 있게 여러가지 환경을 설정해주면서 컴파일 하는 부분
model.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])

# 모델 실행 binary_crossentrop
history = model.fit(X, y, epochs = 100, batch_size = 5)

# 손실함수(loss function)는 떨어지고, 정확도(accuracy)는 높아질수록 좋음. 
# 정확도가 계속 올라가다가 떨어지는 순간이 있음. 이 시점 전에 끊어줘야 함.
```
