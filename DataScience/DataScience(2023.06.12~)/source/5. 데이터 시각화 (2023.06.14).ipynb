<h2>matplotlib 라이브버리로 그래프 그리기</h2>
1. 전체 그래프가 위치할 기본 틀을 만들고<br>
2. 그래프를 그려넣을 격자<br>
3. 격자에 그래프를 하나씩 추가, 순서는 왼쪽->오른쪽<br> 
4. 1행이 차면 2번째 행에 그려넣는다.

```python
dataset_2 = anscombe[anscombe['dataset'] == 'II'] 
dataset_3 = anscombe[anscombe['dataset'] == 'III'] 
dataset_4 = anscombe[anscombe['dataset'] == 'IV'] 
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(2,2,1)  #2행 2열의 첫번째   #행의 크기, 열의크기. #왼쪽상단 1
axes2 = fig.add_subplot(2,2,2)   #오른쪽 상단 2  
axes3 = fig.add_subplot(2,2,3)   #왼쪽 하단 3
axes4 = fig.add_subplot(2,2,4)   #오른쪽 하단4
```

```python
axes1.plot(dataset_1['x'], dataset_1['y'], 'o')
axes2.plot(dataset_2['x'], dataset_2['y'], 'o')
axes3.plot(dataset_3['x'], dataset_3['y'], 'o')
axes4.plot(dataset_4['x'], dataset_4['y'], 'o')

fig    #그래프를 확인하려면 fig
```

```python
#fig(기본틀), 그래프 격자(axes1)
fig = plt.figure()
axes1 = fig.add_subplot(1,1,1)
```

<h3>2. 산점도 그래프</h3>
변수 2개 사용한 '이변량 그래프' - total_bill 열에 따른 tip 열의 분포를 나타낸 산점도 그래프

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'], tips['tip'])
axes1.set_title('Scatterplot of Total Bill cs Tip')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

<h3>3. 박스 그래프 </h3>
    <li>이산형 변수- Femail, Male 처럼 명확하게 구분되는 값을 의미 </li>
    <li>연속형변수 - Tips와 같이 명확하게 셀 수 없는 변위의 값을 의미</li>

### outlier를 어떻게 처리할지 결정할 줄 알아야. outlier는 평균값을 올리는 이상치
### https://devhwi.tistory.com/11


```python
boxplot = plt.figure()
axes1 = boxplot.add_subplot(1, 1, 1)

axes1.boxplot(
    [
        tips[tips['sex'] == 'Female']['tip'],   #tips df에서 성별이 'Female' 인 tip 열
        tips[tips['sex'] == 'Male']['tip'],   #tips df에서 성별이 'Male' 인 tip 열
    ],
    labels = ['Femal', 'Male']
)
axes1.set_xlabel('Sex')
axes1.set_ylabel('Tips')
axes1.set_title('Boxplot of Tips by Sex')
```

<h3>다변량 데이터로 다변량 그래프 그리기 - 산점도 그래프</h3>
<li> 다변량 - 3개 이상의 변수를 사용</li>

```python
#성별을 새 변수로 추가
#(문자열은 산점도 그래프의 색상을 지정하는 값을 사용할 수 없다.)

def recode_sex(sex):
    if sex =='Female':
        return 0
    else:
        return 1
```

```python
#tips df에 sex_color 열을 만든다.
#apply메서드는 내가 만든 함수인 recode_sex를 모드 적용하겠다. 

tips['sex_color'] = tips['sex'].apply(recode_sex)
tips
```

```python
scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x = tips['total_bill'],
    y = tips['tip'],
    s = tips['size'] * 10,
    c = tips['sex_color'],   
    alpha = 0.5      #alpha는 투명도
)

axes1.set_title('Totla Bill cs Tip colored by Sex and Sized by Size')
axes1.set_xlabel('Total Bill')
axes1.set_ylabel('Tip')
```

<h3>단변량 데이터로 그래프그리기</h3>
<li>seaborn 라이브러리는 matplotlib라이브러리를 기반으로 만든 라이브러리

```python
import seaborn as sns
tips = sns.load_dataset('tips')
```

<h3>seaborn 라이브러리로 히스토그램</h3>
<li>seaborn 라이브러리로 히스토그램을 그리려면 subplots,distplot메서드를 사용
<li>distplot 메서드: 기본틀 만듬 / total_bill 메서드: total_bill 열데이터를 전달


```python
import warnings
warnings.filterwarnings(action='ignore')
```

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'])
ax.set_title('Total Bill Histogram with Density Plot')
```

```python
#밀집도(정규화시켜 넓이가 1이 되도록 그린 그래프) 제외
#Kernel Density Estimation(커널밀도추정)
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], kde = False)    #밀도표시하지 않음
ax.set_title('Total Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], hist = False)   
ax.set_title('Total Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Unit Probability')
```

```python
#rug 인자 추가 - 그래프의 축에 동일한 길이의 직선을 붙여 데이터의 밀집정도를 표현한 그래프
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'], rug=True)   #밑에깔린 양탄자모양. 밀도
ax.set_title('Total Bill Histogram with Density and Rug Plot' )
ax.set_xlabel('Total Bill')
```

<h3> count 그래프 - 이산값을 나타낸 그래프</h3>
<li> 주로 범주형 변수의 분포를 파악할 때 사용
<li> sns.countplot(x or y, data)

```python
ax = plt.subplots()
ax = sns.countplot(x = 'day', data=tips )   
ax.set_title('Count of days')
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Frequency')
```

```python
ax = plt.subplots()
ax = sns.countplot(y = 'day', data=tips )   #위와 비교했을 때 x와 y를 바꿈
ax.set_title('Count of days')
ax.set_xlabel('Day of the Week')
ax.set_ylabel('Frequency')
```

<h3>다양한 종류의 이변량 그래프 그리기</h3>
<li>regplot 메서드를 사용하면 산점도 그래프와 회귀선을 그릴 수 있다. 

```python
ax = plt.subplots()
ax = sns.regplot(x = 'total_bill', y = 'tip', data=tips )  
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

```python
ax = plt.subplots()
ax = sns.regplot(x = 'total_bill', y = 'tip', data=tips, fit_reg = False )  #데이터만 나타나도록
ax.set_title('Scatterplot of Total Bill and Tip')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Tip')
```

<h3>5. 바 그래프 그리기</h3>

```python
#barplot은 지정한 변수의 평균을 계산하여 그린다. 
#시간에 따라 지불한 비용의 평균을 바 그래프로 나타냄
ax = plt.subplots()
ax = sns.barplot(x='time', y='total_bill', data=tips)
ax.set_title('Bar plot of average total bill for time of day')
ax.set_xlabel('Time of day')
ax.set_ylabel('Average total bill')
```

<h3>6. 박스그래프 그리기</h3>

### 1. 전체 승객을 나이별로 히스토그램으로 나타내라

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
```

```python
axes1 = fig.add_subplot(1,1,1)
axes1
```

```python
axes1.hist(titanic['age'], bins=100)   #bins=10: x축의 간격을 10으로 조정
axes1.set_title('Histogram of Titanic passenger\'s age')
axes1.set_xlabel('age')
axes1.set_ylabel('num')
fig
```

### 2. 남, 여 승객수를 나타내라(countplot)

```python
ax = plt.subplots()
ax = sns.countplot(x = 'sex', data=titanic )   
ax.set_title('Count of sex')
ax.set_xlabel('성별')
ax.set_ylabel('num')
ax
```

### 3. 사망자와 생존자를 출력해 보세요. (pie)

```python
labels = '사망자', '생존자'
survived = titanic.groupby('survived').size()
dead = titanic.groupby('survived').size()[0]
alive = titanic.groupby('survived').size()[1]
sizes = [dead, alive]
explode = (0, 0.1)   
fig1, ax1 = plt.subplots()
ax1.set_title('Histogram of Titanic Survived rate')
ax1.pie(sizes, explode = explode, labels = labels, autopct = '%1.1f%%', shadow = True, startangle = 90)
ax1.axis('equal')
plt.show()
```
