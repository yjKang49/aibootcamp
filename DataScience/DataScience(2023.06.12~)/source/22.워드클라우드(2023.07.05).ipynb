```python
from wordcloud import WordCloud, STOPWORDS
from nltk.corpus import stopwords
import numpy as np
import matplotlib.pyplot as plt
import platform
from PIL import Image     # 이미지를 처리하는 라이브러리 import
```

```python
#글자 깨짐 방지
from matplotlib import font_manager, rc   
path = 'C:\Windows\Fonts\malgun.ttf'
if platform.system() == 'Darwin':
    rc('font', family = 'AppleGothic')
elif platform.system() == 'Windows':
    font_name = font_manager.FontProperties(fname = path).get_name()
    rc('font', family = font_name)
else:
    print('Unknown system...')
```

```python
text = open('../data/python_data/e_keyw.txt').read() 
stopwords = set(STOPWORDS)
```

```python
wc = WordCloud(background_color = 'white', max_words = 2000,    #최대 2000단어
              colormap = 'summer', margin = 1, random_state = 1).generate(text) 

plt.figure(figsize = (12,12))
plt.imshow(wc)   # interpolation = 'bilinear' : 그림이 겹쳐보이지 않게
plt.axis('off')
plt.show()
```

```python
mask = np.array(Image.open('../img/heart.png'))
wc1 = WordCloud(background_color = 'white', max_words = 2000,    #최대 2000단어
              mask = mask, stopwords = stopwords)
wc1 = wc1.generate(text)   # 앨리스라는 문서에서 단어를 뽑아온다

plt.figure(figsize = (12,8))
plt.imshow(wc1)
plt.axis('off')
plt.show()
```

```python
import nltk
from konlpy.tag import Twitter
t = Twitter()
text_k = open('../data/python_data/k_keyw.txt', encoding = 'utf-8').read() 
tokens_ko = t.nouns(text_k)
tokens_ko
```

```python
ko = nltk.Text(tokens_ko)
plt.figure(figsize = (12,6))
ko.plot(50)
plt.show()   # 불용어를 처리해야할 필요가 있다. 
```

```python
stop_words = ['.', '(', ')', ',', "'", '%', '-', 'X', ').', '×', '린',
              '의','자','에','안','번','호','을','이','다','만',
              '로','가','를','발','인','액','수','세','중','제','월','곳',
             '위','이하','것','표','명', '및']

ko = [each_word for each_word in ko if each_word not in stop_words ]
ko
```

```python
ko = nltk.Text(ko)
plt.figure(figsize = (12,6))
ko.plot(50)     # Plot sorted frequency of top 50 tokens
plt.show()
```

```python
data = ko.vocab().most_common(150)
mask = np.array(Image.open('../img/heart.png'))

wordcloud = WordCloud(font_path = 'C:\Windows\Fonts\malgun.ttf',
                      relative_scaling = 0.1, 
                      mask = mask,
                      background_color = 'white',
                      margin = 1, random_state = 1
                     ).generate_from_frequencies(dict(data))

plt.figure(figsize = (12,8))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
