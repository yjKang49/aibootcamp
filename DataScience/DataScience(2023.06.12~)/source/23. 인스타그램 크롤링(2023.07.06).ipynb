```python
def select_first(driver):
    first = driver.find_element(By.CSS_SELECTOR, 'div._aabd')    #classëŠ” ì¤‘ë³µê°’ì´ìˆê¸° ë•Œë¬¸ì— ì˜¤ë¥˜ë‚˜ë©´ ì—¬ëŸ¬ í´ë˜ìŠ¤ë¥¼ ê°€ì ¸ì™€ì•¼í•¨.
    first.click()                                                #ì—¬ëŸ¬ ì‚¬ì§„ div tagë“¤ì´ ìˆëŠ” í˜ì´ì§€
    time.sleep(10)
```

```python
select_first(driver)
```

```python
# ê²Œì‹œê¸€ ì •ë³´ ê°€ì ¸ì˜¤ê¸°
import re     #ì •ê·œí‘œí˜„ì‹
from bs4 import BeautifulSoup
import unicodedata

def get_content(driver):
    #í˜„ì¬ í˜ì´ì§€ html ì •ë³´ ê°€ì ¸ì˜¤ê¸°
    html = driver.page_source 
    soup = BeautifulSoup(html, 'html.parser')
    #ë³¸ë¬¸ ë‚´ìš© ê°€ì ¸ì˜¤ê¸°
    try:
        content = soup.select('div._a9zs')[0].text
        content = unicodedata.normalize('NFC', content)
    except:
        content = ' '
    # ë³¸ë¬¸ ë‚´ìš©ì—ì„œ í•´ì‹œíƒœê·¸ ê°€ì ¸ì˜¤ê¸°(ì •ê·œì‹ í™œìš©)
    tags = re.findall(r'#[^\s#,\\]+', content)
    #ì‘ì„±ì¼ì ì •ë³´ ê°€ì ¸ì˜¤ê¸°
    date = soup.select('time._aaqe')[0]['datetime'][:10]   #timeíƒœê·¸ì˜ classê°€_aaqeì´ë©´ì„œ datetime attributeì˜ valueë¥¼ 10ê¹Œì§€ ê°€ì ¸ì™€
    #ìˆ˜ì§‘í•œ ì •ë³´ ì €ì¥í•˜ê¸°
    data = [content, date, tags]
    return data

get_content(driver)
```

### 1.5 ë‹¤ìŒ ê²Œì‹œê¸€ ì—´ê¸°

```python
def move_next(driver):
    right = driver.find_element(By.CSS_SELECTOR, 'div._aaqg._aaqh')
    right.click()
    time.sleep(3)
move_next(driver)
```

### 1.6 ì •ë³´ ìˆ˜ì§‘í•˜ê¸°

```python
word = 'ì œì£¼ë„ë§›ì§‘'    #ê²€ìƒ‰ì–´
url = insta_searching(word)
driver.get(url)
time.sleep(10)

select_first(driver)    #ì²« ë²ˆì§¸ ê²Œì‹œê¸€ ì—´ê¸°

results = []
target = 10
for i in range(target):
    try:
        data = get_content(driver)
        results.append(data)
        move_next(driver)
    except:
        time.sleep(5)
        move_next(driver)
        
print(results[:2])
```
